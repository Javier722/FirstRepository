1. Importación de Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. Carga del dataset

```python

df = pd.read_csv("E:/Natsu E/Data science/Project/dataset_ventas_para_limpieza.csv") 
```

3. Creacion de la copia del dataframe con el fin de evitar resultados inesperados

```python
data = df.copy()
```

4. Revisión de los datos

```python
data
```

```python
#Filas duplicadas
duplicados_completos = data.duplicated().sum() 
print(duplicados_completos)
```

5. Limpieza

        5.1 Limpieza de duplicados

```python
#Limpieza de duplicados en la columna "Sale_ID"
data_sin_duplicados = data.drop_duplicates(["Sale_ID"]) 
print(data_sin_duplicados)
```

```python
print(data_sin_duplicados.describe())
```

    5.2 Limpieza de datos vacios

```python
data_sin_NA = data_sin_duplicados.dropna() 
```

```python
print(data_sin_NA)
```

    5.3 Limpieza de valores atipicos o extraños

```python
#Quitamos los valores negativos de la columna 'Quantity_Sold'
data_sin_atipicos = data_sin_NA[data_sin_NA['Quantity_Sold'] > 0] 
```

```python
data_sin_atipicos
```

```python
data_sin_atipicos.info()
```

6. Creación de la Columna "Total_Income"

```python
#Creacion de columna Total Income, usando la funcion ".loc" para prevenir errores 
data_sin_atipicos.loc[:, "Total_Income"] = data_sin_atipicos["Quantity_Sold"] * data_sin_atipicos["Unit_Price"] 
```

```python
data_sin_atipicos
```

```python
#Renombramos el df
data_con_IngresosT = data_sin_atipicos 
```

```python
#Creacion de una copia para evitar posibles errores
data_con_IngresosT = data_con_IngresosT.copy()

#Redondeamos a dos decimales la columna 'Total_Income'
data_con_IngresosT['Total_Income'] = data_con_IngresosT['Total_Income'].round(2) 
```

```python
data_con_IngresosT
```

7. Exportación a archivo .CSV

```python
#Exportacion del archivo a csv para su visualizacion en Power BI
Archivo_corregido = "E:/Natsu E/Data science/Project/dataset_ventas_corregido.csv" 
data_con_IngresosT.to_csv(Archivo_corregido, index=False)
print(f"Archivo corregido guardado en: {Archivo_corregido}")
```
