```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv("E:/Natsu E/Data science/Project/dataset_ventas_para_limpieza.csv") #Carga del dataset
```

```python
data
```

```python
duplicados_completos = data.duplicated().sum() #Filas duplicadas
print(duplicados_completos)
```

```python
data_sin_duplicados = data.drop_duplicates(["Sale_ID"]) #Limpieza de duplicados en la columna "Sale_ID"
print(data_sin_duplicados)
```

```python
print(data_sin_duplicados.describe())
```

```python
data_sin_NA = data_sin_duplicados.dropna() #Limpieza de datos vacios
```

```python
print(data_sin_NA)
```

```python
data_sin_atipicos = data_sin_NA[data_sin_NA['Quantity_Sold'] > 0] #Limpieza de valores extraÃ±os o atipicos como valores negativos
```

```python
data_sin_atipicos
```

```python
data_sin_atipicos.info()
```

```python
data_sin_atipicos.loc[:, "Total_Income"] = data_sin_atipicos["Quantity_Sold"] * data_sin_atipicos["Unit_Price"] #Creacion de columna Total Income, usando la funcion ".loc" para prevenir errores 
```

```python
data_sin_atipicos
```

```python
data_con_IngresosT = data_sin_atipicos
```

```python
data_con_IngresosT = data_con_IngresosT.copy() #creacion de una copia para evitar posibles errores
data_con_IngresosT['Total_Income'] = data_con_IngresosT['Total_Income'].round(2) #redondeamos a dos decimales la columna 'Total_Income'
```

```python
data_con_IngresosT
```

```python
Archivo_corregido = "E:/Natsu E/Data science/Project/dataset_ventas_corregido.csv" #Exportacion del archivo a csv para su visualizacion en Power BI
data_con_IngresosT.to_csv(Archivo_corregido, index=False)
print(f"Archivo corregido guardado en: {Archivo_corregido}")
```
